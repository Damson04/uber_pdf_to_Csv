```python
import os
import tabula
import pdfplumber
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
folder_path = r'C:\Users\samso\OneDrive\Desktop\metro college Data\PROJECT\uber job\uberstatement\2024'  # Your folder
all_data = []

# Regex to extract key parts from transaction lines
transaction_pattern = re.compile(
    r'(?P<processed>\w{3}, \w{3} \d{1,2})\s+'
    r'(?P<event>UberX|UberX Share|Comfort|Courier|Tips|Transferred To Bank Account)\s+'
    r'(?P<earnings>-?CA\$[\d,]+\.\d{2})?\s*'
    r'(?P<refunds>-?CA\$[\d,]+\.\d{2})?\s*'
    r'(?P<balance>CA\$[\d,]+\.\d{2})',
    re.IGNORECASE
)

# Loop through each PDF file
for file_name in os.listdir(folder_path):
    if file_name.endswith('.pdf'):
        file_path = os.path.join(folder_path, file_name)
        print(f"Processing: {file_path}")
        
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                text = page.extract_text()
                if not text:
                    continue
                
                lines = text.split('\n')
                for line in lines:
                    match = transaction_pattern.search(line)
                    if match:
                        row = match.groupdict()
                        row['source_file'] = file_name
                        all_data.append(row)

# Create DataFrame
df_2024 = pd.DataFrame(all_data)

# Clean and convert currency strings to float
for col in ['earnings', 'refunds', 'balance']:
    df_2024[col] = df_2024[col].str.replace('CA\$', '', regex=True).str.replace(',', '').astype(float)

# Rename columns for clarity
df_2024.rename(columns={
    'processed': 'Processed',
    'event': 'Event',
    'earnings': 'Your earnings',
    'refunds': 'Refunds & Expenses',
    'balance': 'Balance'
}, inplace=True)

# Show and export
print(df.head())
df_2024.to_csv("uber2024_transactions_extracted.csv", index=False)
print(" Saved as 'uber2024_transactions_extracted.csv'")
```
